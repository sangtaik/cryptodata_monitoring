```python
import json
import pandas as pd

# jupyter 실행
import nest_asyncio
import asyncio

exchange_name = "korbit"
SYMBOL = "btc_krw"
asks_list = []
bids_list = []

# 파일 경로 지정
filepath = "orderbook_data/orderbook_btc_krw_20250404_133210.jsonl"  # 너의 실제 저장 파일로 바꿔줘

with open(filepath, "r") as f:
    for line in f:
        record = json.loads(line)
        timestamp = record["timestamp"]
        order_data = record.get("data", {}).get("data", {})
        asks = order_data.get("asks", [])[:10]
        bids = order_data.get("bids", [])[:10]

        for ask in asks:
            asks_list.append({
                "timestamp": timestamp,
                "side": "ask",
                "price": float(ask["price"]),
                "qty": float(ask["qty"])
            })

        for bid in bids:
            bids_list.append({
                "timestamp": timestamp,
                "side": "bid",
                "price": float(bid["price"]),
                "qty": float(bid["qty"])
            })

# 합치기
df = pd.DataFrame(asks_list + bids_list)

# 확인
#print(df.head())
print(df.to_string())
```
